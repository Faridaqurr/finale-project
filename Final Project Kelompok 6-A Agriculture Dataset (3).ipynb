```python
import matplotlib.pyplot as plt
import pandas as pd

# Data statistik deskriptif
data = {
    'Date From': {
        'count': 330.000000,
        'mean': -1536.809091,
        'std': 2867.255946,
        'min': -10000.000000,
        '25%': -3000.000000,
        '50%': -800.000000,
        '75%': 930.000000,
        'max': 2000.000000
    },
    'Date To': {
        'count': 330.000000,
        'mean': 1059.727273,
        'std': 1447.224197,
        'min': -7000.000000,
        '25%': 710.000000,
        '50%': 1810.000000,
        '75%': 1914.250000,
        'max': 2001.000000
    }
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menentukan kolom yang akan diplot (misalnya, mean, 50%, max)
kolom_diplot = ['mean', '50%', 'max']

# Membuat grafik batang
plt.figure(figsize=(10, 5))
df.T[kolom_diplot].plot(kind='bar', alpha=0.7)
plt.title('Statistik Deskriptif Date From dan Date To')
plt.xlabel('Statistik')
plt.ylabel('Nilai')
plt.legend(loc='upper left')
plt.show()
```

Distribusi data pada kolom "Date From" dan "Date To" memiliki karakteristik simetris, karena nilai mean dan median cenderung berada pada posisi yang hampir sama. Distribusi simetris tidak selalu menandakan bahwa distribusi tersebut adalah distribusi normal. Karena, tidak semua distribusi simetris adalah distribusi normal.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data statistik deskriptif
data = {
    'Date From': {
        'count': 330.000000,
        'mean': -1536.809091,
        'std': 2867.255946,
        'min': -10000.000000,
        '25%': -3000.000000,
        '50%': -800.000000,
        '75%': 930.000000,
        'max': 2000.000000
    },
    'Date To': {
        'count': 330.000000,
        'mean': 1059.727273,
        'std': 1447.224197,
        'min': -7000.000000,
        '25%': 710.000000,
        '50%': 1810.000000,
        '75%': 1914.250000,
        'max': 2001.000000
    }
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Kolom yang akan diplot
kolom_diplot = ['min', '25%', '50%', '75%', 'max']

# Membuat grafik bar
plt.figure(figsize=(10, 5))
df.T[kolom_diplot].plot(kind='bar', alpha=0.7, yerr=df.T['std'])
plt.title('Range dan Standar Deviasi Date From dan Date To')
plt.xlabel('Statistik')
plt.ylabel('Nilai')
plt.legend(loc='upper left')
plt.show()
```

- Dalam 'Date From', rentang data terlihat cukup besar, dengan nilai minimum yang sangat rendah dan nilai maksimum yang jauh lebih tinggi. Ini menunjukkan adanya variasi yang signifikan dalam rentang waktu 'Date From'. Standar deviasi yang tinggi juga menegaskan variasi yang besar dalam data ini, menandakan bahwa data tersebar luas dari nilai rata-rata.

- Di sisi lain, 'Date To' memiliki rentang data yang juga cukup luas, meskipun tidak sebesar 'Date From'. Standar deviasi yang lebih rendah pada 'Date To' mengindikasikan bahwa data cenderung lebih terkumpul di sekitar nilai rata-rata. Median dan kuartil bawah dan atas yang lebih tinggi pada 'Date To' menunjukkan bahwa sebagian besar data berada di bagian atas rentang.

- Secara keseluruhan, perbandingan antara 'Date From' dan 'Date To' menunjukkan perbedaan yang signifikan dalam variasi dan sebaran data. 'Date From' memiliki variasi yang lebih besar dan rentang yang lebih luas, sementara 'Date To' lebih terkumpul di sekitar nilai rata-rata dengan variasi yang lebih terkendali.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
sns.lineplot(data=Agriculture_Dataset, y="Date From", x='NGA', color='blue')
plt.xticks(rotation=90)
plt.title('Hubungan antara time dan NGA', fontsize=18, fontweight="bold")
```

Grafik tersebut menunjukkan bahwa praktik pertanian dapat ditelusuri kembali hingga sekitar 10000 Sebelum Masehi

```python
plt.figure(figsize=(30,30))
g = sns.catplot(x="NGA", y="Date To", kind="box",
            data=Agriculture_Dataset.sort_values("NGA"), palette="tab10")
g.fig.set_figwidth(15)
g.fig.set_figheight(8)
plt.title('NGA vs Date To',fontsize=15, fontweight='bold')
plt.xticks(rotation=90, fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(30,30))
g = sns.catplot(x="Variable", y="Date From", kind="boxen",
            data=Agriculture_Dataset.sort_values("NGA"), palette="tab10")
g.fig.set_figwidth(12)
g.fig.set_figheight(8)
plt.title('NGA vs Date From',fontsize=15, fontweight='bold')
plt.xticks(rotation=90, fontsize=14)
plt.show()
```

Dari grafik - grafik tersebut, dapat dilihat bahwa :
- Karbohidrat dikonsumsi sejak sekitar 10000 SM.
- Pupuk diadopsi sebagai bagian dari pertanian sejak sekitar 7500 SM.
- Irigasi juga mulai digunakan dalam pertanian sejak 8000 SM.
- Sistem tanam juga diadopsi sekitar 10000 SM.

```python
Agriculture_Dataset_irrigation = Agriculture_Dataset[(Agriculture_Dataset.Variable == 'Irrigation Coefficient')]
Agriculture_Dataset_irrigation.head(5)
Agriculture_Dataset_irrigation[['Value From', 'Value To']] = Agriculture_Dataset_irrigation[['Value From', 'Value To']].apply(pd.to_numeric)
```

```python
plt.figure(figsize=(20,15))
sns.barplot(data=Agriculture_Dataset_irrigation, x="NGA", y="Value From", palette="viridis")
plt.xticks(fontsize=15, rotation = 90)
plt.yticks(fontsize=15)
plt.title('Coefficient of Irrigation across NGAs', fontsize=18, fontweight='bold')
```

- Deccan, Kachi Plain, Kansai Plain, Susiana, dan Yemeni Coastal Plain memiliki koefisien irigasi tertinggi.
- Latium dan Sogdiana memiliki koefisien irigasi sebesar 1, menunjukkan bahwa terdapat curah hujan yang cukup tinggi di wilayah ini dan irigasi tidak memberikan perbedaan signifikan dalam produktivitas tanaman.
- Dari 30 NGA unik yang tersedia, hanya ada 15 di antaranya yang menggunakan irigasi pada setiap waktu tertentu. Hal ini mungkin terkait dengan kondisi cuaca di berbagai NGA.

```python
Agriculture_Dataset_fertilizer = Agriculture_Dataset[(Agriculture_Dataset.Variable == 'Fertilizer Coefficient')]
Agriculture_Dataset_fertilizer.head(5)
Agriculture_Dataset_fertilizer[['Value From', 'Value To']] = Agriculture_Dataset_fertilizer[['Value From', 'Value To']].apply(pd.to_numeric)
```

```python
plt.figure(figsize=(20,15))
sns.barplot(data=Agriculture_Dataset_fertilizer, x="NGA", y="Value From", palette="terrain")
plt.xticks(fontsize=15, rotation = 90)
plt.yticks(fontsize=15)
plt.title('Coefficient of Fertilizer across NGAs', fontsize=18, fontweight='bold')
```

- 23 dari 30 di antaranya menggunakan pupuk dalam pertanian mereka.
- Koefisien Pupuk tertinggi sekitar 1,90 ada di Ghanian Coast.
- Koefisien Deccan, Konya Plain, dan Paris Basin bervariasi dari waktu ke waktu. Hal ini mungkin disebabkan oleh jenis sumber pupuk yang digunakan pada berbagai waktu.
- Hampir semua koefisien pupuk lebih besar dari 1, menunjukkan bahwa penggunaan pupuk membantu meningkatkan produktivitas tanaman.
